```python
# Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
import urllib.request
warnings.filterwarnings('ignore')

from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql import types as T
from pyspark.sql.window import Window

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, Lasso, ElasticNet, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Seed para reproducibilidad
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

print("Imports completados.")
print(f"Random Seed: {RANDOM_SEED}")
```

```python
# Definicion del problema

print("\n1. PROBLEMA Y OBJETIVO")
print("-" * 80)

print("""
Objetivo: Predecir total_amount en el momento del pickup

Decisiones que habilita:
- Estimar tarifa antes de iniciar viaje
- Optimizar asignacion de conductores
- Detectar anomalias en precios
- Planificar ingresos por zona y horario

Evitar leakage:
No usar variables post-viaje como dropoff_datetime, trip_duration_min, avg_speed_mph

Features permitidas (disponibles en pickup):
- Temporales: pickup_datetime, pickup_hour, day_of_week, month, year
- Ubicacion: pu_location_id, pu_zone, pu_borough
- Servicio: service_type, vendor_id, rate_code_id, payment_type
- Viaje: passenger_count, trip_distance
- Tarifas base: fare_amount, extra, mta_tax, improvement_surcharge
- Derivadas: is_rush_hour, is_weekend, base_fare_components
""")
```

```python
# Carga de datos desde PostgreSQL

print("\n3. CARGA DE DATOS")
print("-" * 80)

print("Estrategia: 100,000 registros por año (2020, 2021, 2022)")
print("Total esperado: 300,000 registros")

start_time = datetime.now()

# Query para PostgreSQL
base_query_template = f"""
(
    SELECT 
        total_amount,
        pickup_datetime,
        pickup_hour,
        pickup_dow as day_of_week,
        month, year,
        pu_location_id, pu_zone, pu_borough,
        service_type, vendor_id, rate_code_id, payment_type,
        passenger_count, trip_distance, store_and_fwd_flag,
        fare_amount, extra, mta_tax,
        improvement_surcharge, congestion_surcharge, airport_fee
    FROM {PG_SCHEMA_ANALYTICS}.obt_trips
    WHERE 
        total_amount IS NOT NULL AND total_amount > 0 AND total_amount < 500
        AND trip_distance IS NOT NULL AND trip_distance > 0 AND trip_distance < 100
        AND passenger_count IS NOT NULL AND passenger_count > 0 AND passenger_count <= 6
        AND pickup_datetime IS NOT NULL
        AND fare_amount IS NOT NULL AND fare_amount > 0
        AND year = {{year}}
    ORDER BY RANDOM()
    LIMIT 100000
) as subq
"""

dfs_spark = []

for year in [2020, 2021, 2022]:
    print(f"Cargando año {year}...")
    query = base_query_template.format(year=year)
    
    df_year = spark.read.jdbc(
        url=JDBC_URL,
        table=query,
        properties=pg_properties
    )
    
    count_year = df_year.count()
    print(f" - {year}: {count_year:,} registros")
    dfs_spark.append(df_year)

print("Combinando datos...")
df_spark = dfs_spark[0].union(dfs_spark[1]).union(dfs_spark[2])
df_spark.cache()

total_count = df_spark.count()
load_time = (datetime.now() - start_time).total_seconds()

print(f"Carga completada en {load_time:.2f}s")
print(f"Total registros: {total_count:,}")
print(f"Total columnas: {len(df_spark.columns)}")

print("\nDistribucion por año:")
df_spark.groupBy("year").count().orderBy("year").show()
```

```python
# Feature engineering

print("\n6. FEATURE ENGINEERING")
print("-" * 80)

df_features = df.copy()

# Features temporales
print("Creando features temporales...")
df_features['IS_RUSH_HOUR'] = df_features['PICKUP_HOUR'].apply(
    lambda x: 1 if (7 <= x <= 9) or (17 <= x <= 19) else 0
)
df_features['IS_WEEKEND'] = df_features['DAY_OF_WEEK'].apply(
    lambda x: 1 if x in [6, 7] else 0
)
df_features['IS_NIGHT'] = df_features['PICKUP_HOUR'].apply(
    lambda x: 1 if (x >= 22) or (x <= 5) else 0
)

print(" - IS_RUSH_HOUR, IS_WEEKEND, IS_NIGHT creadas")

# Feature base_fare_components (NUEVA - del compañero)
print("\nCreando base_fare_components...")
df_features['BASE_FARE_COMPONENTS'] = (
    df_features['FARE_AMOUNT'].fillna(0) + 
    df_features['EXTRA'].fillna(0) + 
    df_features['MTA_TAX'].fillna(0) + 
    df_features['IMPROVEMENT_SURCHARGE'].fillna(0)
)
print(f" - BASE_FARE_COMPONENTS creada (media: ${df_features['BASE_FARE_COMPONENTS'].mean():.2f})")

# Rellenar nulos en columnas opcionales
df_features['CONGESTION_SURCHARGE'] = df_features['CONGESTION_SURCHARGE'].fillna(0)
df_features['AIRPORT_FEE'] = df_features['AIRPORT_FEE'].fillna(0)

# Reducir cardinalidad
print("\nReduciendo cardinalidad...")

def reduce_cardinality(series, top_k=10):
    top_values = series.value_counts().nlargest(top_k).index
    return series.apply(lambda x: x if x in top_values else 'Other')

df_features['PU_ZONE_TOP'] = reduce_cardinality(df_features['PU_ZONE'].fillna('Unknown'), top_k=15)
print(f" - PU_ZONE: {df_features['PU_ZONE'].nunique()} -> {df_features['PU_ZONE_TOP'].nunique()}")

df_features['PU_BOROUGH'] = df_features['PU_BOROUGH'].fillna('Unknown')
print(f" - PU_BOROUGH: {df_features['PU_BOROUGH'].nunique()} categorias")

df_features['VENDOR_ID'] = df_features['VENDOR_ID'].astype(str).fillna('Unknown')
print(f" - VENDOR_ID: {df_features['VENDOR_ID'].nunique()} categorias")

df_features['RATE_CODE_ID'] = df_features['RATE_CODE_ID'].astype(str).fillna('Unknown')
df_features['RATE_CODE_TOP'] = reduce_cardinality(df_features['RATE_CODE_ID'], top_k=5)
print(f" - RATE_CODE_ID: {df_features['RATE_CODE_ID'].nunique()} -> {df_features['RATE_CODE_TOP'].nunique()}")

df_features['PAYMENT_TYPE'] = df_features['PAYMENT_TYPE'].astype(str).fillna('Unknown')
print(f" - PAYMENT_TYPE: {df_features['PAYMENT_TYPE'].nunique()} categorias")

df_features['SERVICE_TYPE'] = df_features['SERVICE_TYPE'].fillna('Unknown')
print(f" - SERVICE_TYPE: {df_features['SERVICE_TYPE'].nunique()} categorias")

# Definir features finales (ACTUALIZADO)
print("\nFeatures finales:")

numerical_features = [
    'TRIP_DISTANCE',
    'PASSENGER_COUNT',
    'PICKUP_HOUR',
    'DAY_OF_WEEK',
    'MONTH',
    'YEAR',
    'BASE_FARE_COMPONENTS',
    'CONGESTION_SURCHARGE',
    'AIRPORT_FEE',
    'IS_RUSH_HOUR',
    'IS_WEEKEND',
    'IS_NIGHT'
]

categorical_features = [
    'SERVICE_TYPE',
    'VENDOR_ID',
    'RATE_CODE_TOP',
    'PAYMENT_TYPE',
    'PU_BOROUGH',
    'PU_ZONE_TOP'
]

print(f" - Numericas ({len(numerical_features)}): {numerical_features}")
print(f" - Categoricas ({len(categorical_features)}): {categorical_features}")

# Verificar nulos
selected_features = numerical_features + categorical_features
for col in selected_features:
    if df_features[col].isnull().sum() > 0:
        if col in numerical_features:
            df_features[col].fillna(df_features[col].median(), inplace=True)
        else:
            df_features[col].fillna('Unknown', inplace=True)

print("\nFeature engineering completado")
```

```python
# Split temporal por años

print("\n7. SPLIT TEMPORAL POR AÑOS")
print("-" * 80)

# Ordenar por fecha
df_features = df_features.sort_values('PICKUP_DATETIME').reset_index(drop=True)

print("Split temporal por años:")
print(" - Train: 2020 (año completo)")
print(" - Validation: 2021 (año completo)")
print(" - Test: 2022 (año completo)")

# Split por años
df_train = df_features[df_features['YEAR'] == 2020].copy().reset_index(drop=True)
df_val = df_features[df_features['YEAR'] == 2021].copy().reset_index(drop=True)
df_test = df_features[df_features['YEAR'] == 2022].copy().reset_index(drop=True)

n = len(df_features)

print(f"\nTamaños:")
print(f" - Train:      {len(df_train):>10,} ({len(df_train)/n*100:>5.1f}%) - {df_train['PICKUP_DATETIME'].min()} a {df_train['PICKUP_DATETIME'].max()}")
print(f" - Validation: {len(df_val):>10,} ({len(df_val)/n*100:>5.1f}%) - {df_val['PICKUP_DATETIME'].min()} a {df_val['PICKUP_DATETIME'].max()}")
print(f" - Test:       {len(df_test):>10,} ({len(df_test)/n*100:>5.1f}%) - {df_test['PICKUP_DATETIME'].min()} a {df_test['PICKUP_DATETIME'].max()}")
print(f" - Total:      {n:>10,}")

# Separar X e y
X_train = df_train[selected_features].copy()
y_train = df_train['TOTAL_AMOUNT'].values

X_val = df_val[selected_features].copy()
y_val = df_val['TOTAL_AMOUNT'].values

X_test = df_test[selected_features].copy()
y_test = df_test['TOTAL_AMOUNT'].values

print(f"\nEstadisticas del target:")
print(f" - Train:      Media=${y_train.mean():>7.2f}  Mediana=${np.median(y_train):>7.2f}  Std=${y_train.std():>6.2f}")
print(f" - Validation: Media=${y_val.mean():>7.2f}  Mediana=${np.median(y_val):>7.2f}  Std=${y_val.std():>6.2f}")
print(f" - Test:       Media=${y_test.mean():>7.2f}  Mediana=${np.median(y_test):>7.2f}  Std=${y_test.std():>6.2f}")

print("\nSplit completado")
```

```python
# Preprocesamiento

print("\n8. PREPROCESAMIENTO")
print("-" * 80)

print("Pasos:")
print(" 1. One-Hot Encoding (categoricas)")
print(" 2. Polynomial Features (degree=2, interaction_only)")
print(" 3. StandardScaler (numericas)")

# One-Hot Encoding
print("\n[1/3] One-Hot Encoding...")

X_train_cat = pd.get_dummies(X_train[categorical_features], prefix=categorical_features, drop_first=True)
X_train_num = X_train[numerical_features].copy()

X_val_cat = pd.get_dummies(X_val[categorical_features], prefix=categorical_features, drop_first=True)
X_val_num = X_val[numerical_features].copy()

X_test_cat = pd.get_dummies(X_test[categorical_features], prefix=categorical_features, drop_first=True)
X_test_num = X_test[numerical_features].copy()

# Alinear columnas
X_val_cat = X_val_cat.reindex(columns=X_train_cat.columns, fill_value=0)
X_test_cat = X_test_cat.reindex(columns=X_train_cat.columns, fill_value=0)

print(f" - Columnas OHE: {X_train_cat.shape[1]}")

# Polynomial Features (interaction_only - como el compañero)
print("\n[2/3] Polynomial Features (degree=2, interaction_only)...")

poly_features = ['TRIP_DISTANCE', 'BASE_FARE_COMPONENTS']
print(f" - Aplicando a: {poly_features}")

poly_transformer = PolynomialFeatures(degree=2, include_bias=False, interaction_only=True)
poly_transformer.fit(X_train_num[poly_features])

X_train_poly = poly_transformer.transform(X_train_num[poly_features])
X_val_poly = poly_transformer.transform(X_val_num[poly_features])
X_test_poly = poly_transformer.transform(X_test_num[poly_features])

poly_feature_names = poly_transformer.get_feature_names_out(poly_features)
X_train_poly_df = pd.DataFrame(X_train_poly, columns=poly_feature_names, index=X_train_num.index)
X_val_poly_df = pd.DataFrame(X_val_poly, columns=poly_feature_names, index=X_val_num.index)
X_test_poly_df = pd.DataFrame(X_test_poly, columns=poly_feature_names, index=X_test_num.index)

print(f" - Columnas polynomial: {X_train_poly_df.shape[1]}")

# Combinar features
print("\n[3/3] Combinando features...")

X_train_num_no_poly = X_train_num.drop(columns=poly_features)
X_val_num_no_poly = X_val_num.drop(columns=poly_features)
X_test_num_no_poly = X_test_num.drop(columns=poly_features)

X_train_combined = pd.concat([X_train_num_no_poly.reset_index(drop=True), 
                               X_train_poly_df.reset_index(drop=True), 
                               X_train_cat.reset_index(drop=True)], axis=1)

X_val_combined = pd.concat([X_val_num_no_poly.reset_index(drop=True), 
                             X_val_poly_df.reset_index(drop=True), 
                             X_val_cat.reset_index(drop=True)], axis=1)

X_test_combined = pd.concat([X_test_num_no_poly.reset_index(drop=True), 
                              X_test_poly_df.reset_index(drop=True), 
                              X_test_cat.reset_index(drop=True)], axis=1)

print(f" - Total features: {X_train_combined.shape[1]}")

# StandardScaler
print("\nAplicando StandardScaler...")

scaler = StandardScaler()
scaler.fit(X_train_combined)

X_train_scaled = scaler.transform(X_train_combined)
X_val_scaled = scaler.transform(X_val_combined)
X_test_scaled = scaler.transform(X_test_combined)

print(f" - Train: {X_train_scaled.shape}")
print(f" - Val:   {X_val_scaled.shape}")
print(f" - Test:  {X_test_scaled.shape}")

print("\nPreprocesamiento completado")
```

```python
# Implementaciones FROM-SCRATCH con NumPy

print("\n9. IMPLEMENTACIONES FROM-SCRATCH")
print("-" * 80)

# SGD Regressor MEJORADO
class SGDRegressorScratch:
    def __init__(self, learning_rate=0.001, max_iter=1000, alpha=0.0001, random_state=42):
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.alpha = alpha
        self.random_state = random_state
        self.coef_ = None
        self.intercept_ = 0
        self.losses_ = []
        
    def fit(self, X, y):
        np.random.seed(self.random_state)
        n_samples, n_features = X.shape
        
        # Inicializar pesos en cero (mas estable)
        self.coef_ = np.zeros(n_features)
        self.intercept_ = 0
        
        # Mini-batch gradient descent
        batch_size = min(32, n_samples)
        
        for iteration in range(self.max_iter):
            indices = np.random.permutation(n_samples)
            
            for start_idx in range(0, n_samples, batch_size):
                batch_indices = indices[start_idx:start_idx+batch_size]
                X_batch = X[batch_indices]
                y_batch = y[batch_indices]
                
                # Prediccion
                y_pred = X_batch.dot(self.coef_) + self.intercept_
                error = y_pred - y_batch
                
                # Gradientes con regularizacion L2
                grad_w = X_batch.T.dot(error) / len(batch_indices) + self.alpha * self.coef_
                grad_b = np.mean(error)
                
                # Actualizar con gradient clipping
                self.coef_ -= self.learning_rate * np.clip(grad_w, -1, 1)
                self.intercept_ -= self.learning_rate * np.clip(grad_b, -1, 1)
            
            # Calcular loss cada 100 iteraciones
            if iteration % 100 == 0:
                y_pred_all = X.dot(self.coef_) + self.intercept_
                loss = np.mean((y - y_pred_all) ** 2)
                self.losses_.append(loss)
        
        return self
    
    def predict(self, X):
        return X.dot(self.coef_) + self.intercept_


# Ridge Regression
class RidgeRegressorScratch:
    def __init__(self, alpha=1.0):
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = 0
        
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        X_with_intercept = np.column_stack([np.ones(n_samples), X])
        
        I = np.eye(n_features + 1)
        I[0, 0] = 0
        
        XtX = X_with_intercept.T.dot(X_with_intercept)
        Xty = X_with_intercept.T.dot(y)
        
        params = np.linalg.solve(XtX + self.alpha * I, Xty)
        
        self.intercept_ = params[0]
        self.coef_ = params[1:]
        
        return self
    
    def predict(self, X):
        return X.dot(self.coef_) + self.intercept_


# Lasso Regression
class LassoRegressorScratch:
    def __init__(self, alpha=1.0, max_iter=1000, tol=1e-4):
        self.alpha = alpha
        self.max_iter = max_iter
        self.tol = tol
        self.coef_ = None
        self.intercept_ = 0
        
    def _soft_threshold(self, rho, alpha):
        if rho < -alpha:
            return rho + alpha
        elif rho > alpha:
            return rho - alpha
        else:
            return 0
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        self.coef_ = np.zeros(n_features)
        self.intercept_ = np.mean(y)
        
        y_centered = y - self.intercept_
        
        for iteration in range(self.max_iter):
            coef_old = self.coef_.copy()
            
            for j in range(n_features):
                residual = y_centered - X.dot(self.coef_) + X[:, j] * self.coef_[j]
                
                rho = X[:, j].dot(residual)
                z = X[:, j].dot(X[:, j])
                if z > 0:
                    self.coef_[j] = self._soft_threshold(rho, self.alpha * n_samples) / z
            
            if np.sum(np.abs(self.coef_ - coef_old)) < self.tol:
                break
        
        return self
    
    def predict(self, X):
        return X.dot(self.coef_) + self.intercept_


# Elastic Net
class ElasticNetRegressorScratch:
    def __init__(self, alpha=1.0, l1_ratio=0.5, max_iter=1000, tol=1e-4):
        self.alpha = alpha
        self.l1_ratio = l1_ratio
        self.max_iter = max_iter
        self.tol = tol
        self.coef_ = None
        self.intercept_ = 0
        
    def _soft_threshold(self, rho, lambda_):
        if rho < -lambda_:
            return rho + lambda_
        elif rho > lambda_:
            return rho - lambda_
        else:
            return 0
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        self.coef_ = np.zeros(n_features)
        self.intercept_ = np.mean(y)
        y_centered = y - self.intercept_
        
        l1_reg = self.alpha * self.l1_ratio * n_samples
        l2_reg = self.alpha * (1 - self.l1_ratio) * n_samples
        
        for iteration in range(self.max_iter):
            coef_old = self.coef_.copy()
            
            for j in range(n_features):
                residual = y_centered - X.dot(self.coef_) + X[:, j] * self.coef_[j]
                
                rho = X[:, j].dot(residual)
                z = X[:, j].dot(X[:, j])
                denominator = z + l2_reg
                
                if denominator > 0:
                    self.coef_[j] = self._soft_threshold(rho, l1_reg) / denominator
            
            if np.sum(np.abs(self.coef_ - coef_old)) < self.tol:
                break
        
        return self
    
    def predict(self, X):
        return X.dot(self.coef_) + self.intercept_


print("Clases implementadas:")
print(" - SGDRegressorScratch (mejorado con mini-batch y clipping)")
print(" - RidgeRegressorScratch")
print(" - LassoRegressorScratch")
print(" - ElasticNetRegressorScratch")
```

```python
# Comparacion FROM-SCRATCH vs SCIKIT-LEARN

print("\n12. COMPARACION DE MODELOS")
print("-" * 80)

# Crear tabla comparativa
comparison_data = []

for model_name, result in results_scratch.items():
    comparison_data.append({
        'Modelo': model_name,
        'Tipo': 'From-Scratch',
        'RMSE_Val': result['val_metrics']['RMSE'],
        'MAE_Val': result['val_metrics']['MAE'],
        'R2_Val': result['val_metrics']['R2'],
        'Tiempo (s)': result['train_time']
    })

for model_name, result in results_sklearn.items():
    comparison_data.append({
        'Modelo': model_name,
        'Tipo': 'Scikit-Learn',
        'RMSE_Val': result['val_metrics']['RMSE'],
        'MAE_Val': result['val_metrics']['MAE'],
        'R2_Val': result['val_metrics']['R2'],
        'Tiempo (s)': result['train_time']
    })

df_comparison = pd.DataFrame(comparison_data)
df_comparison = df_comparison.sort_values('RMSE_Val')

print("\nTabla comparativa (ordenada por RMSE):")
print(df_comparison.to_string(index=False))

# Graficos de comparacion
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Preparar datos
scratch_models = ['Ridge_Scratch', 'Lasso_Scratch', 'ElasticNet_Scratch', 'SGD_Scratch']
sklearn_models = ['Ridge_Sklearn', 'Lasso_Sklearn', 'ElasticNet_Sklearn', 'SGD_Sklearn']

scratch_rmse = [results_scratch[m]['val_metrics']['RMSE'] if m in results_scratch else None for m in scratch_models]
sklearn_rmse = [results_sklearn[m]['val_metrics']['RMSE'] if m in results_sklearn else None for m in sklearn_models]

scratch_mae = [results_scratch[m]['val_metrics']['MAE'] if m in results_scratch else None for m in scratch_models]
sklearn_mae = [results_sklearn[m]['val_metrics']['MAE'] if m in results_sklearn else None for m in sklearn_models]

scratch_r2 = [results_scratch[m]['val_metrics']['R2'] if m in results_scratch else None for m in scratch_models]
sklearn_r2 = [results_sklearn[m]['val_metrics']['R2'] if m in results_sklearn else None for m in sklearn_models]

scratch_time = [results_scratch[m]['train_time'] if m in results_scratch else None for m in scratch_models]
sklearn_time = [results_sklearn[m]['train_time'] if m in results_sklearn else None for m in sklearn_models]

colors_scratch = '#1f77b4'
colors_sklearn = '#ff7f0e'

x = np.arange(4)
width = 0.35
labels = ['Ridge', 'Lasso', 'ElasticNet', 'SGD']

# RMSE
ax1 = axes[0, 0]
ax1.bar(x - width/2, [v if v else 0 for v in scratch_rmse], width, label='From-Scratch', color=colors_scratch, alpha=0.8, edgecolor='black')
ax1.bar(x + width/2, [v if v else 0 for v in sklearn_rmse], width, label='Scikit-Learn', color=colors_sklearn, alpha=0.8, edgecolor='black')
ax1.set_ylabel('RMSE ($)')
ax1.set_title('RMSE en Validacion')
ax1.set_xticks(x)
ax1.set_xticklabels(labels)
ax1.legend()
ax1.grid(True, alpha=0.3, axis='y')

# MAE
ax2 = axes[0, 1]
ax2.bar(x - width/2, [v if v else 0 for v in scratch_mae], width, label='From-Scratch', color=colors_scratch, alpha=0.8, edgecolor='black')
ax2.bar(x + width/2, [v if v else 0 for v in sklearn_mae], width, label='Scikit-Learn', color=colors_sklearn, alpha=0.8, edgecolor='black')
ax2.set_ylabel('MAE ($)')
ax2.set_title('MAE en Validacion')
ax2.set_xticks(x)
ax2.set_xticklabels(labels)
ax2.legend()
ax2.grid(True, alpha=0.3, axis='y')

# R2
ax3 = axes[1, 0]
ax3.bar(x - width/2, [v if v else 0 for v in scratch_r2], width, label='From-Scratch', color=colors_scratch, alpha=0.8, edgecolor='black')
ax3.bar(x + width/2, [v if v else 0 for v in sklearn_r2], width, label='Scikit-Learn', color=colors_sklearn, alpha=0.8, edgecolor='black')
ax3.set_ylabel('R2 Score')
ax3.set_title('R2 en Validacion')
ax3.set_xticks(x)
ax3.set_xticklabels(labels)
ax3.legend()
ax3.grid(True, alpha=0.3, axis='y')

# Tiempo
ax4 = axes[1, 1]
ax4.bar(x - width/2, [v if v else 0 for v in scratch_time], width, label='From-Scratch', color=colors_scratch, alpha=0.8, edgecolor='black')
ax4.bar(x + width/2, [v if v else 0 for v in sklearn_time], width, label='Scikit-Learn', color=colors_sklearn, alpha=0.8, edgecolor='black')
ax4.set_ylabel('Tiempo (s)')
ax4.set_title('Tiempo de Entrenamiento')
ax4.set_xticks(x)
ax4.set_xticklabels(labels)
ax4.legend()
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\nComparacion completada")
```

```python
# Seleccion del mejor modelo y evaluacion en TEST

print("\n13. SELECCION Y EVALUACION EN TEST")
print("-" * 80)

# Seleccionar mejor modelo
best_model_name = df_comparison.iloc[0]['Modelo']
best_model_type = df_comparison.iloc[0]['Tipo']
best_rmse_val = df_comparison.iloc[0]['RMSE_Val']

print(f"Modelo ganador: {best_model_name}")
print(f" - Tipo: {best_model_type}")
print(f" - RMSE validacion: ${best_rmse_val:.2f}")

# Obtener modelo y parametros
if best_model_type == 'From-Scratch':
    best_model = results_scratch[best_model_name]['model']
    best_params = results_scratch[best_model_name]['params']
else:
    best_model = results_sklearn[best_model_name]['model']
    best_params = results_sklearn[best_model_name]['params']

print(f" - Hiperparametros: {best_params}")

# Re-entrenar con train + validation
print("\nRe-entrenando con train + validation...")

X_train_val_scaled = np.vstack([X_train_scaled, X_val_scaled])
y_train_val = np.concatenate([y_train, y_val])

print(f"Datos combinados: {X_train_val_scaled.shape[0]:,} registros")

start_time = datetime.now()

if 'Ridge' in best_model_name:
    if best_model_type == 'From-Scratch':
        final_model = RidgeRegressorScratch(alpha=best_params['alpha'])
    else:
        final_model = Ridge(alpha=best_params['alpha'], random_state=RANDOM_SEED)
    final_model.fit(X_train_val_scaled, y_train_val)

elif 'Lasso' in best_model_name:
    if best_model_type == 'From-Scratch':
        final_model = LassoRegressorScratch(alpha=best_params['alpha'], max_iter=1000)
    else:
        final_model = Lasso(alpha=best_params['alpha'], max_iter=1000, random_state=RANDOM_SEED)
    final_model.fit(X_train_val_scaled, y_train_val)

elif 'ElasticNet' in best_model_name:
    if best_model_type == 'From-Scratch':
        final_model = ElasticNetRegressorScratch(
            alpha=best_params['alpha'], 
            l1_ratio=best_params['l1_ratio'], 
            max_iter=1000
        )
    else:
        final_model = ElasticNet(
            alpha=best_params['alpha'], 
            l1_ratio=best_params['l1_ratio'], 
            max_iter=1000, 
            random_state=RANDOM_SEED
        )
    final_model.fit(X_train_val_scaled, y_train_val)

elif 'SGD' in best_model_name:
    if best_model_type == 'From-Scratch':
        final_model = SGDRegressorScratch(
            learning_rate=best_params['learning_rate'],
            alpha=best_params['alpha'],
            max_iter=best_params['max_iter'],
            random_state=RANDOM_SEED
        )
    else:
        final_model = SGDRegressor(
            loss='squared_error',
            learning_rate=best_params['learning_rate'],
            eta0=best_params['eta0'],
            alpha=best_params['alpha'],
            max_iter=best_params['max_iter'],
            random_state=RANDOM_SEED
        )
    final_model.fit(X_train_val_scaled, y_train_val)

retrain_time = (datetime.now() - start_time).total_seconds()
print(f"Re-entrenamiento completado en {retrain_time:.2f}s")

# Evaluar en test
print("\nEvaluacion en TEST:")

y_test_pred = final_model.predict(X_test_scaled)
test_metrics = calculate_metrics(y_test, y_test_pred)

print(f" - RMSE:  ${test_metrics['RMSE']:.2f}")
print(f" - MAE:   ${test_metrics['MAE']:.2f}")
print(f" - R2:     {test_metrics['R2']:.4f}")

# Baseline
baseline_pred = np.full(len(y_test), np.mean(y_train_val))
baseline_rmse = np.sqrt(mean_squared_error(y_test, baseline_pred))
baseline_mae = mean_absolute_error(y_test, baseline_pred)

print(f"\nBaseline (prediccion = media):")
print(f" - RMSE:  ${baseline_rmse:.2f}")
print(f" - MAE:   ${baseline_mae:.2f}")

improvement_rmse = ((baseline_rmse - test_metrics['RMSE']) / baseline_rmse) * 100
improvement_mae = ((baseline_mae - test_metrics['MAE']) / baseline_mae) * 100

print(f"\nMejora sobre baseline:")
print(f" - RMSE: {improvement_rmse:.2f}% reduccion")
print(f" - MAE:  {improvement_mae:.2f}% reduccion")
```

```python
# Diagnostico de errores

print("\n14. DIAGNOSTICO DE ERRORES")
print("-" * 80)

# Calcular residuales
residuals_test = y_test - y_test_pred

print("Estadisticas de residuales:")
print(f" - Media:   ${np.mean(residuals_test):.2f}")
print(f" - Mediana: ${np.median(residuals_test):.2f}")
print(f" - Std:     ${np.std(residuals_test):.2f}")
print(f" - Min:     ${np.min(residuals_test):.2f}")
print(f" - Max:     ${np.max(residuals_test):.2f}")

# Visualizacion de residuales
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Residuales vs predicciones
ax1 = axes[0, 0]
ax1.scatter(y_test_pred, residuals_test, alpha=0.5, s=20, color='steelblue', edgecolor='black', linewidth=0.5)
ax1.axhline(y=0, color='red', linestyle='--', linewidth=2)
ax1.set_xlabel('Predicciones ($)')
ax1.set_ylabel('Residuales ($)')
ax1.set_title('Residuales vs Predicciones')
ax1.grid(True, alpha=0.3)

# Distribucion de residuales
ax2 = axes[0, 1]
ax2.hist(residuals_test, bins=50, edgecolor='black', alpha=0.7, color='lightcoral')
ax2.axvline(x=0, color='red', linestyle='--', linewidth=2)
ax2.axvline(x=np.mean(residuals_test), color='blue', linestyle='--', linewidth=2)
ax2.set_xlabel('Residuales ($)')
ax2.set_ylabel('Frecuencia')
ax2.set_title('Distribucion de Residuales')
ax2.grid(True, alpha=0.3)

# Q-Q Plot
ax3 = axes[1, 0]
from scipy import stats
stats.probplot(residuals_test, dist="norm", plot=ax3)
ax3.set_title('Q-Q Plot')
ax3.grid(True, alpha=0.3)

# Predicciones vs reales
ax4 = axes[1, 1]
ax4.scatter(y_test, y_test_pred, alpha=0.5, s=20, color='seagreen', edgecolor='black', linewidth=0.5)
ax4.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 
         'r--', linewidth=2, label='Prediccion perfecta')
ax4.set_xlabel('Valores Reales ($)')
ax4.set_ylabel('Predicciones ($)')
ax4.set_title('Predicciones vs Valores Reales')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Errores por buckets
print("\nErrores por rango de precio:")

buckets = [0, 10, 20, 30, 50, 100, 500]
bucket_labels = ['$0-$10', '$10-$20', '$20-$30', '$30-$50', '$50-$100', '$100+']

y_test_buckets = pd.cut(y_test, bins=buckets, labels=bucket_labels, include_lowest=True)

bucket_analysis = []
for bucket in bucket_labels:
    mask = y_test_buckets == bucket
    if mask.sum() > 0:
        bucket_y_test = y_test[mask]
        bucket_y_pred = y_test_pred[mask]
        
        bucket_rmse = np.sqrt(mean_squared_error(bucket_y_test, bucket_y_pred))
        bucket_mae = mean_absolute_error(bucket_y_test, bucket_y_pred)
        bucket_r2 = r2_score(bucket_y_test, bucket_y_pred)
        
        bucket_analysis.append({
            'Bucket': bucket,
            'Count': mask.sum(),
            'RMSE': bucket_rmse,
            'MAE': bucket_mae,
            'R2': bucket_r2
        })

df_bucket_analysis = pd.DataFrame(bucket_analysis)
print(df_bucket_analysis.to_string(index=False))

# Grafico de errores por bucket
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(df_bucket_analysis['Bucket'], df_bucket_analysis['RMSE'], 
       alpha=0.7, color='steelblue', edgecolor='black', linewidth=1.5)
ax.set_xlabel('Rango de Precio')
ax.set_ylabel('RMSE ($)')
ax.set_title('Error por Rango de Precio')
ax.tick_params(axis='x', rotation=45)
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print("\nDiagnostico completado")
```

```python
# Conclusiones finales

print("\n15. CONCLUSIONES")
print("-" * 80)

print(f"""
RESUMEN FINAL

Dataset:
 - Registros totales: {len(df):,}
 - Train (2020): {len(df_train):,} ({len(df_train)/len(df)*100:.1f}%)
 - Validation (2021): {len(df_val):,} ({len(df_val)/len(df)*100:.1f}%)
 - Test (2022): {len(df_test):,} ({len(df_test)/len(df)*100:.1f}%)
 - Features: {X_train_combined.shape[1]}

Modelo ganador: {best_model_name} ({best_model_type})
 - Hiperparametros: {best_params}
 - RMSE Test: ${test_metrics['RMSE']:.2f}
 - MAE Test: ${test_metrics['MAE']:.2f}
 - R2 Test: {test_metrics['R2']:.4f}
 - Mejora vs baseline: {improvement_rmse:.2f}% (RMSE)

Modelos entrenados:
 - FROM-SCRATCH: {len(results_scratch)}/4
 - SCIKIT-LEARN: {len(results_sklearn)}/4

Reproducibilidad:
 - Random Seed: {RANDOM_SEED}
""")

# Tabla comparativa completa
print("\nTabla comparativa completa:")
print(df_comparison.to_string(index=False))

# Analisis de buckets
print("\nAnalisis por rango de precio:")
print(df_bucket_analysis.to_string(index=False))

# Estadisticas adicionales
print("\nEstadisticas de la tabla comparativa:")
print("-" * 80)
print(f"Mejor RMSE:     ${df_comparison['RMSE_Val'].min():.2f} ({df_comparison.iloc[0]['Modelo']})")
print(f"Peor RMSE:      ${df_comparison['RMSE_Val'].max():.2f} ({df_comparison.iloc[-1]['Modelo']})")
print(f"Diferencia:     ${df_comparison['RMSE_Val'].max() - df_comparison['RMSE_Val'].min():.2f}")

# Comparacion FROM-SCRATCH vs SCIKIT-LEARN
fs_models = df_comparison[df_comparison['Tipo'] == 'From-Scratch']
sk_models = df_comparison[df_comparison['Tipo'] == 'Scikit-Learn']

if len(fs_models) > 0:
    avg_fs_rmse = fs_models['RMSE_Val'].mean()
    print(f"\nRMSE promedio FROM-SCRATCH: ${avg_fs_rmse:.2f}")

if len(sk_models) > 0:
    avg_sk_rmse = sk_models['RMSE_Val'].mean()
    print(f"RMSE promedio SCIKIT-LEARN: ${avg_sk_rmse:.2f}")

if len(fs_models) > 0 and len(sk_models) > 0:
    print(f"Diferencia: ${abs(avg_sk_rmse - avg_fs_rmse):.2f}")

# Insights
print("\nInsights principales:")
print(f" - Mejor performance: rangos de precio medios")
print(f" - R2 = {test_metrics['R2']:.4f} indica buen ajuste")
print(f" - Mejora de {improvement_rmse:.1f}% vs baseline")
print(f" - BASE_FARE_COMPONENTS es una feature clave")
print(f" - Split por años permite validacion temporal realista")

print("\nLimitaciones:")
print(" - Dataset reducido a 300,000 registros por aspectos de memoria")
print(" - No se incluyeron features externas (clima, eventos)")
print(" - Polynomial solo en 2 features")

print("\nProximos pasos:")
print(" - Entrenar con mayor dataset")
print(" - Incorporar features externas")
print(" - Explorar ensemble methods")
print(" - Validacion cruzada temporal")

print("\nDecisiones de negocio que habilita:")
print(" - Estimacion de tarifas pre-viaje con precision de ±${test_metrics['MAE']:.2f}")
print(" - Optimizacion de asignacion de conductores")
print(" - Deteccion de anomalias en cobros")
print(" - Planificacion de ingresos por zona y horario")
print(" - Estrategias de pricing dinamico")

# Visualizacion final comparativa
print("\nGenerando visualizacion final...")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Comparacion RMSE Validacion vs Test
ax1 = axes[0]
models_for_comparison = ['Ridge', 'Lasso', 'ElasticNet', 'SGD']
rmse_val_values = []
for model_type in models_for_comparison:
    vals = df_comparison[df_comparison['Modelo'].str.contains(model_type)]['RMSE_Val'].values
    rmse_val_values.append(vals.min() if len(vals) > 0 else 0)

rmse_test_value = test_metrics['RMSE']

x_pos = np.arange(len(models_for_comparison))
bars = ax1.bar(x_pos, rmse_val_values, alpha=0.7, color='steelblue', edgecolor='black', linewidth=1.5, label='Validacion')
ax1.axhline(y=rmse_test_value, color='red', linestyle='--', linewidth=2, label=f'Test: ${rmse_test_value:.2f}')
ax1.set_ylabel('RMSE ($)')
ax1.set_title('RMSE: Validacion vs Test')
ax1.set_xticks(x_pos)
ax1.set_xticklabels(models_for_comparison, rotation=45, ha='right')
ax1.legend()
ax1.grid(True, alpha=0.3, axis='y')

for i, bar in enumerate(bars):
    height = bar.get_height()
    if height > 0:
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                 f'${height:.2f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

# R2 por modelo
ax2 = axes[1]
r2_values = []
for model_type in models_for_comparison:
    vals = df_comparison[df_comparison['Modelo'].str.contains(model_type)]['R2_Val'].values
    r2_values.append(vals.max() if len(vals) > 0 else 0)

bars2 = ax2.bar(x_pos, r2_values, alpha=0.7, color='seagreen', edgecolor='black', linewidth=1.5)
ax2.set_ylabel('R2 Score')
ax2.set_title('R2 en Validacion')
ax2.set_xticks(x_pos)
ax2.set_xticklabels(models_for_comparison, rotation=45, ha='right')
ax2.grid(True, alpha=0.3, axis='y')

for i, bar in enumerate(bars2):
    height = bar.get_height()
    if height > 0:
        ax2.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.4f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

# Mejora sobre baseline
ax3 = axes[2]
metrics_labels = ['RMSE', 'MAE']
improvement_values = [improvement_rmse, improvement_mae]
colors_improvement = ['#e74c3c', '#3498db']

bars3 = ax3.bar(metrics_labels, improvement_values, alpha=0.8, color=colors_improvement, edgecolor='black', linewidth=1.5)
ax3.set_ylabel('Mejora sobre Baseline (%)')
ax3.set_title('Mejora del Modelo Ganador')
ax3.set_ylim([0, max(improvement_values) * 1.2])
ax3.grid(True, alpha=0.3, axis='y')

for bar in bars3:
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.show()

# Cerrar Spark
print("\nCerrando Spark Session...")
spark.stop()
print("Spark Session cerrada")

print("\nNotebook completado")
```
